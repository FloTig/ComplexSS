```python
import math
import random
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation
from Belousov import Belousov
```

## Calculating Parameter space alpha

```python
for i in np.arange(0.7,1.0,0.01):
    bel = Belousov(i,1,1,250,1500)
    A = bel.A
    B= bel.B
    C = bel.C
    grid = bel.grid
    time = np.linspace(0,bel.t,bel.t)
    cutoff =0
    plt.plot(time,A,label = 'A')
    plt.plot(time,B,label = 'B')
    plt.plot(time,C,label = 'C')
    plt.xlabel('Time (t)')
    plt.ylabel('Concentration')
    plt.title(f'Alpha {i}')
    plt.legend()
    plt.savefig(f'Alpha{i}.png')
    plt.close()

    plt.plot(A[cutoff:],B[cutoff:],'k-',linewidth =.25, label = 'phase space')
    plt.xlabel('A')
    plt.ylabel('B')
    plt.legend()
    plt.savefig(f'PhaseAlphaAB{i}.png')
    plt.close()

    plt.plot(B[cutoff:],C[cutoff:],'k-',linewidth =.25, label = 'phase space')
    plt.xlabel('B')
    plt.ylabel('C')
    plt.legend()
    plt.savefig(f'PhaseAlphaBC{i}.png')
    plt.close()


    plt.plot(A[cutoff:],C[cutoff:],'k-',linewidth =.25, label = 'phase space')
    plt.xlabel('B')
    plt.ylabel('C')
    plt.legend()
    plt.savefig(f'PhaseAlphaAC{i}.png')
    plt.close()
```
