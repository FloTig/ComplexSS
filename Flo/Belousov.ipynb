```python
import math
import random
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation
from Belousov import Belousov
```

```python
bel2 = Belousov(1,0.6,1,250,50)
A2 = bel2.A
B2 = bel2.B
C2 = bel2.C
grid = bel2.grid
time = np.arange(0,bel2.t,bel2.dt)
```

```python
c1 = 0
c2 = 100
plt.plot(time[c1:c2],A2[c1:c2],label = 'A')
plt.plot(time[c1:c2],B2[c1:c2],label = 'B')
plt.plot(time[c1:c2],C2[c1:c2],label = 'C')
plt.xlabel('Time (t)')
plt.ylabel('Concentration (mol)')
plt.legend()
plt.show()
```

```python
cutoff = 0
plt.plot(A2[cutoff:],B2[cutoff:],'k-',linewidth =.25, label = 'phase space')
plt.xlabel('A')
plt.ylabel('B')
plt.legend()
plt.show()
```

```python

plt.plot(A2[cutoff:],C2[cutoff:],'k-',linewidth =.25, label = 'phase space')
plt.xlabel('A')
plt.ylabel('C')
plt.legend()
plt.show()
```

```python

plt.plot(B2[cutoff:],C2[cutoff:],'k-',linewidth =.25, label = 'phase space')
plt.xlabel('B')
plt.ylabel('C')
plt.legend()
plt.show()
```
